```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
############################### appartements-meubles #############################
```

```python
for p in range(1,99): 
    #url
    url = f"https://www.expat-dakar.com/appartements-a-louer?page={p}"
    # recuperer le code html
    res = requests.get(url)
    #creer l'objet beautifulsoup
    soup = bs(res.text, "html.parser")
    #v1: details 
    details = soup.find_all('div', class_="listings-cards__list")
    for detail in details:
        print(detail.text) 
    #V2 : nombre chambre
    nb_chambres = soup.find_all("span", class_="listing-card__header__tags__item listing-card__header__tags__item--no-of-bedrooms listing-card__header__tags__item--no-of-bedrooms_3")
    total = 0
    for nb_chambre in nb_chambres:
        print(nb_chambre.text)
        total += int(nb_chambre.text[0])
    print(total)
    nb_chambrs = soup.find_all("span", class_="listing-card__header__tags__item listing-card__header__tags__item--no-of-bedrooms listing-card__header__tags__item--no-of-bedrooms_1")
    total = 0
    for nb_chambr in nb_chambrs:
        print(nb_chambr.text)
        total += int(nb_chambr.text[0])
    print(total)
    nb_chambs = soup.find_all("span", class_="listing-card__header__tags__item listing-card__header__tags__item--no-of-bedrooms listing-card__header__tags__item--no-of-bedrooms_2")

    total = 0
    for nb_chamb in nb_chambs:
        print(nb_chamb.text)
        total += int(nb_chamb.text[0])
    print(total)
    #V3 : superficie
    superficies = soup.find_all("span", class_="listing-card__header__tags__item listing-card__header__tags__item--square-metres listing-card__header__tags__item--square-metres_150")
    for superficie in superficies:
        print(superficie.text)
    superficies1= soup.find_all("span",class_="listing-card__header__tags__item listing-card__header__tags__item--square-metres listing-card__header__tags__item--square-metres_80")
    for superficie1 in superficies1:
        print(superficie1.text)
    superficies2= soup.find_all("span",class_="listing-card__header__tags__item listing-card__header__tags__item--square-metres listing-card__header__tags__item--square-metres_80")
    for superficie2 in superficies2:
        print(superficie2.text)
    superficies3 = soup.find_all("span", class_="listing-card__header__tags__item listing-card__header__tags__item--square-metres listing-card__header__tags__item--square-metres_150")
    for superficie3 in superficies3:
        print(superficie3.text)
    superficies4 = soup.find_all("span", class_="listing-card__header__tags__item listing-card__header__tags__item--square-metres listing-card__header__tags__item--square-metres_200")
    for superficie4 in superficies4:
        print(superficie4.text)
    superficies5 = soup.find_all("span", class_="listing-card__header__tags__item listing-card__header__tags__item--square-metres listing-card__header__tags__item--square-metres_50")
    for superficie5 in superficies5:
        print(superficie5.text)
    superficies6 = soup.find_all("span", class_="listing-card__header__tags__item listing-card__header__tags__item--square-metres listing-card__header__tags__item--square-metres_25")
    for superficie6 in superficies6:
        print(superficie6.text)
    superficies7 = soup.find_all("span", class_="listing-card__header__tags__item listing-card__header__tags__item--square-metres listing-card__header__tags__item--square-metres_100")
    for superficie7 in superficies7:
        print(superficie7.text)
    superficies8 = soup.find_all("span", class_="listing-card__header__tags__item listing-card__header__tags__item--square-metres listing-card__header__tags__item--square-metres_130")
    for superficie8 in superficies8:
        print(superficie8.text)
    superficies9 = soup.find_all("span", class_="listing-card__header__tags__item listing-card__header__tags__item--square-metres listing-card__header__tags__item--square-metres_146")
    for superficie9 in superficies9:
        print(superficie9.text)
    #V4 : adresse
    adresses = soup.find_all("div", class_="listing-card__header__location")
    for adresse in adresses:
        print(adresse.text)
    #V5 : prix
    prix = soup.find_all("span", class_="listing-card__price")
    for i in prix:
        print(i.text)
    #v6:image_lien
    images = soup.find_all("div", class_="listing-card__image__inner")
    for image in images:
        print(image.img["src"])
    
```
